# FUNCTIONS II

- A function of the form $f(x) = mx + b$, for constants **m** and **b**, is called a **linear function**.
- A function $f(x) = x^a$, where **a** is a constant, is called a **power function**.
- A function **P** is a **polynomial** if $P(x) = a_nx^n + a_{n-1}x^{n-1} + ... + a_1x + a_0$, where n is a nonnegative integer and the numbers $a_0,a_1,a_2,a_3,...$ are real constants (called the **coefficients of the polynomial**), **n** is called the **degree of the polynomial**. Polynomials of **degree 2**, usually written as $P(x) = ax^2 + bx + c$ are called **quadratic functions**.
- A **rational function** is a quotient or ratio of two polynomials: $\dfrac{P(x)}{Q(x)}$
- Functions of the form $f(x) = a^x$, where the base $a > 0$ is is a positive constant and $a \neq 1$, are called **exponential functions**.
- A function of the form $f(x) = \int_a^x$, where $a \neq 1$, is called a **logarithmic function**.
- An **algebraic function** is a function constructed from polynomials using algebraic operations (**addition**, **subtraction**, **multiplication**, **division**, and **takingroots**).

# Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import random,math
from warnings import filterwarnings
filterwarnings("ignore",category=UserWarning)
filterwarnings("ignore",category=RuntimeWarning)
```

```python
plt.style.use("dark_background")
```

```python
class RESULTS:
    pass
```

```python
def PLOTFUNCTION(x:list or float or int,
                 y:list or float or int,
                 title:str)->RESULTS:
    fig = plt.figure(figsize=(15,8))
    ax = fig.add_subplot(1,1,1)
    ax.spines["left"].set_position("zero")
    ax.spines["bottom"].set_position("zero")
    ax.spines["right"].set_position("zero")
    ax.spines["top"].set_position("zero")
    ax.spines["right"].set_color("white")
    ax.spines["top"].set_color("white")
    ax.set_aspect('equal',adjustable='datalim')
    plt.title(str(title).upper())
    plt.plot(x,y)
    plt.tight_layout()
    plt.show()
```

```python
def MULTIPLOTFUNCTION(x:list or float or int,
                      y_1:list or float or int,
                      y_2:list or float or int,
                      y_3:list or float or int,
                      title:str)->RESULTS:
    fig = plt.figure(figsize=(15,8))
    ax = fig.add_subplot(1,1,1)
    ax.spines["left"].set_position("zero")
    ax.spines["bottom"].set_position("zero")
    ax.spines["right"].set_position("zero")
    ax.spines["top"].set_position("zero")
    ax.spines["right"].set_color("white")
    ax.spines["top"].set_color("white")
    ax.set_aspect('equal',adjustable='datalim')
    plt.title(str(title).upper())
    plt.plot(x,y_1,color="green")
    plt.plot(x,y_2,color="yellow")
    plt.plot(x,y_3,color="white")
    plt.tight_layout()
    plt.show()
```

# Linear Functions

```python
xdata = np.linspace(-3,3,1000)
```

```python
def LINEARFUNCTION(x:int or float,
                   m:int=3,
                   b:int=2):
    return (x*m)+b
```

```python
def RANDOMSLOPELINEARFUNCTION(x:int or float):
    rm = random.choice(range(1,5))
    bm = random.choice(range(1,10))
    return (x*int(rm))+bm
```

```python
y_m_3 = []
y_m_2 = []
y_m_1 = []
for _i in xdata:
    y_m_3.append(LINEARFUNCTION(_i))
    y_m_2.append(LINEARFUNCTION(_i,m=2))
    y_m_1.append(LINEARFUNCTION(_i,m=1))
```

```python
PLOTFUNCTION(x,y_m_3,"m = 3")
```

```python
PLOTFUNCTION(x,y_m_2,"m = 2")
```

```python
PLOTFUNCTION(x,y_m_1,"m = 1")
```

```python
MULTIPLOTFUNCTION(x,
                  y_m_3,
                  y_m_2,
                  y_m_1,
                  "m = 3 (green), m = 2 (yellow), m = 1 (white)")
```

```python
y_m_r = []
y_m_b = []
y_m_t = []
for _i in xdata:
    y_m_r.append(RANDOMSLOPELINEARFUNCTION(_i))
    y_m_b.append(RANDOMSLOPELINEARFUNCTION(_i))
    y_m_t.append(RANDOMSLOPELINEARFUNCTION(_i))
```

```python
MULTIPLOTFUNCTION(x,
                  y_m_r,
                  y_m_b,
                  y_m_t,
                  "Random M And B")
```

# Power Functions

```python
xdata = np.linspace(-10,10,500)
```

```python
def POWERFUNCTION(x:int or float,power:int or float):
    return x**(power)
```

```python
y_half = []
y_2 = []
y_3 = []
y_4 = []
for _i in xdata:
    y_half.append(POWERFUNCTION(_i,float(1/2)))
    y_2.append(POWERFUNCTION(_i,2))
    y_3.append(POWERFUNCTION(_i,3))
    y_4.append(POWERFUNCTION(_i,4))
```

```python
PLOTFUNCTION(xdata,y_2,"Power Function, Power = 2")
```

```python
PLOTFUNCTION(xdata,y_3,"Power Function, Power = 3")
```

```python
PLOTFUNCTION(xdata,y_4,"Power Function, Power = 4")
```

```python
PLOTFUNCTION(xdata,y_half,"Power Function, Power = 1/2")
```

# Polynomial Functions

```python
xdata = np.linspace(-2,2,1000)
```

```python
def POLYNOMIALFUNCTION(x:int or float,
                       n:int=4,
                       a:int or float=1):
    a1 = a+5
    a2 = a+3
    a3 = a+1
    p = a1*(x**n) + a2*(x**(n-1)) + a3*(x**(n-2)) + 0
    return p
```

```python
y_5 = []
y_4 = []
y_3 = []
for _i in xdata:
    y_5.append(POLYNOMIALFUNCTION(_i,n=5))
    y_4.append(POLYNOMIALFUNCTION(_i,n=4))
    y_3.append(POLYNOMIALFUNCTION(_i,n=3))
```

```python
PLOTFUNCTION(xdata,y_5,"Polynomial Function, Degree 5")
```

```python
PLOTFUNCTION(xdata,y_4,"Polynomial Function, Degree 4")
```

```python
PLOTFUNCTION(xdata,y_3,"Polynomial Function, Degree 3")
```

# Rational Functions

```python
xdata = np.linspace(-2,2,1000)
```

```python
def RATIONALFUNCTION(p1:list or float or int,
                     q1:list or float or int):
    rf = p1/q1
    return rf
```

```python
p1 = []
q1 = []
rf = []
for _i in xdata:
    p1.append(POLYNOMIALFUNCTION(_i,n=5))
    q1.append(POLYNOMIALFUNCTION(_i,n=3))
for i_,x_ in enumerate(p1):
    v2 = q1[i_]
    rf.append(RATIONALFUNCTION(x_,v2))
```

```python
PLOTFUNCTION(xdata,rf,"Rational Function Example - Power: 5-3 = 2")
```

```python
p1 = []
q1 = []
rf = []
for _i in xdata:
    p1.append(POLYNOMIALFUNCTION(_i,n=4))
    q1.append(POLYNOMIALFUNCTION(_i,n=3))
for i_,x_ in enumerate(p1):
    v2 = q1[i_]
    rf.append(RATIONALFUNCTION(x_,v2))
```

```python
PLOTFUNCTION(xdata,rf,"Rational Function Example - Power: 4-3 = 1")
```

# Exponential Functions

```python
xdata = np.linspace(1,5,1000)
```

```python
def EXPONENTIALFUNCTION(x:int or float or list,
                       a:int or float = 2):
    assert a > 1
    return a**x
```

```python
y_2 = []
y_3 = []
y_5 = []
for _i in xdata:
    y_2.append(EXPONENTIALFUNCTION(_i,a=2))
    y_3.append(EXPONENTIALFUNCTION(_i,a=3))
    y_5.append(EXPONENTIALFUNCTION(_i,a=5))
```

```python
PLOTFUNCTION(xdata,y_2,"EXPONENTIAL FUNCTION - a: 2")
```

```python
PLOTFUNCTION(xdata,y_3,"EXPONENTIAL FUNCTION - a: 3")
```

```python
PLOTFUNCTION(xdata,y_5,"EXPONENTIAL FUNCTION - a: 5")
```

# Logarithmic Functions

```python
xdata = np.linspace(1,15,1000)
```

```python
def LOGARITHMICFUNCTION(x:int or float or list,
                        a:int=10):
    assert a != 1
    lg = math.log(x,a)
    return lg
```

```python
y_10 = []
y_5 = []
y_2 = []
for _i in xdata:
    y_10.append(LOGARITHMICFUNCTION(_i,a=10))
    y_5.append(LOGARITHMICFUNCTION(_i,a=5))
    y_2.append(LOGARITHMICFUNCTION(_i,a=2))
```

```python
PLOTFUNCTION(xdata,y_10,"LOGARITHMIC FUNCTION - Base: 10")
```

```python
PLOTFUNCTION(xdata,y_5,"LOGARITHMIC FUNCTION - Base: 5")
```

```python
PLOTFUNCTION(xdata,y_2,"LOGARITHMIC FUNCTION - Base: 2")
```
